```python
data = []
choice = 'Randomly'

def line():
    print('------------***-------------')

def course_list():
    line()
    print('Course List')
    line()
    print('1. Add Courses')
    print('2. Mark as done')
    print('3. View Course list')
    print('4. Exit')
    line()
    choice = input('Enter your choice: ')
    return choice

while choice != '4':
    choice = course_list()
    if choice == '1':
        course = input('Enter your course name:  ')
        data.append(course)
        print('Course', course , 'added')
    elif choice == '2':
        item = input('Which course do you want to delete? :  ')
        if item in data:
            data.remove(item)
            print('Removed course: ', item)
        else:
            print('Could not find  ', item , 'course')
    elif choice == '3':
        print('List of the courses:')
        for item in data:
            print(item)
    elif choice == '4':
        print('Goodbye!')
    else:
        print('Please enter one of 1, 2, 3 or 4')
```
